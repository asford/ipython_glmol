Read pdb from rcsb and generate a rosetta pose.

```python
with tempfile.NamedTemporaryFile() as tf:
    source_atoms = pdb_parsing.parse_pdb(urllib2.urlopen("http://www.rcsb.org/pdb/download/downloadFile.do?fileFormat=pdb&compression=NO&structureId=1QYS").read())
    tf.write(pdb_parsing.to_pdb(source_atoms))
    
    source_structure = rosetta.pose_from_pdb(tf.name)
    source_structure.pdb_info().name("1QYS")

# Reset pdbinfo to align residue numbering
source_structure.pdb_info( rosetta.core.pose.PDBInfo(source_structure) )
```

Display pose with sensible default representation.

```python
pose_display(source_structure, line="all;heavy")
```

Define representation based on pose data.

```python
def extract_ss_spans(ss_pattern):
    """Extract all contiguous blocks of secondary structure return grouped by span type."""
    ss_pattern = numpy.array(list(ss_pattern))
    
    ss_edges = list(numpy.flatnonzero(ss_pattern[1:] != ss_pattern[:-1]) + 1)
    ss_spans = zip([0] + ss_edges, ss_edges + [-1])
    
    result = {}
    
    for t in numpy.unique(ss_pattern):
        result[t] = [(start + 1, end + 1) for start, end in ss_spans if ss_pattern[start] == t]
    
    return result

def setup_ss_display(target_pose):
    s_display = pose_display(target_pose)
    import rosetta.core.scoring.dssp

    ss_spans = extract_ss_spans( rosetta.core.scoring.dssp.Dssp(target_pose).get_dssp_secstruct())

    s_display.repr_entries["color"].extend("salmon:elem C;residue %s-%s" % span for span in ss_spans["H"])
    s_display.repr_entries["color"].extend("cyan:elem C;residue %s-%s" % span for span in ss_spans["E"])
    s_display.repr_entries["line"].append("all; heavy")
    return s_display
```
